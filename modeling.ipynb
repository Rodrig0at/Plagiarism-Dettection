```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
```

```python
train = "C:\\Users\\rodri\\OneDrive\\Documentos\\8vo\\Plagiarism-Dettection\\train_clean.csv"
val= "C:\\Users\\rodri\\OneDrive\\Documentos\\8vo\\Plagiarism-Dettection\\val_clean.csv"
test = "C:\\Users\\rodri\\OneDrive\\Documentos\\8vo\\Plagiarism-Dettection\\test_clean.csv"
```

```python
train_df = pd.read_csv(train)
val_df = pd.read_csv(val)
test_df = pd.read_csv(test)
```

```python
x_train = train_df['code']
y_train = train_df['uid']

x_val = val_df['code']
y_val = val_df['uid']
```

```python
tfidf_vectorizer = TfidfVectorizer()
x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_val_tfidf = tfidf_vectorizer.transform(x_val)
x_test_tfidf = tfidf_vectorizer.transform(test_df['code'])
```

```python
modelo = LinearSVC(class_weight='balanced', max_iter=10000)
modelo.fit(x_train_tfidf, y_train)
```

```python
y_pred = modelo.predict(x_val_tfidf)
y_test_pred = modelo.predict(x_test_tfidf)
```

```python
print(classification_report(y_val, y_pred))
```

```python
results = pd.DataFrame({
    'pid': test_df['pid'],
    'uid': y_test_pred
})
results.to_csv('predictions.csv', index=False)
```
